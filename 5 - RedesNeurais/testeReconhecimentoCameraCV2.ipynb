```python
import cv2
import numpy as np
from keras.layers import Dense
from keras.models import Sequential

cap = cv2.VideoCapture(0)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

y = []
images = []

while True:    
    ret, frame = cap.read()
    cv2.imshow('webcam', frame)

    k = cv2.waitKey(1)

    frame = np.expand_dims(cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY), axis=0)
    frame = frame / 255

    cap_img = False
    
    if k == ord("q"):
        break

    if k == ord('0'):   
        classe = 0
        cap_img = True

    if k == ord('1'):
        classe = 1
        cap_img = True        

    if k == ord('t'):        
        images = np.array(images)
        y = np.array(y)
        images = images.reshape((len(images), np.prod(images.shape[1:])))
        dimInput = images.shape[1]
        
        model = Sequential()
        model.add(Dense(units = 150, input_dim = dimInput, activation="relu"))
        model.add(Dense(units = 150, activation="relu"))
        model.add(Dense(units = 1, activation="sigmoid"))
        model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])        
        model.fit(images, y.ravel(), epochs=15, shuffle=True)

    if k == ord("p"):
        imagesP = np.array([frame])
        imagesP = imagesP.reshape((len(imagesP), np.prod(imagesP.shape[1:]))) 

        predict = model.predict(imagesP)
        predict_label = list(np.where(predict >= 0.5, 1, 0))
        labels = ["Cartucho", "Bola"]
        print("Rótulo Encontrado", labels[int(predict_label[0])])
        print("Rótulo", predict)

    if cap_img:
        print("Classe selecionada para treino: ", classe)

        images.append(frame)
        y.append(classe)

        print("y = ", y)

cap.release()
cv2.destroyAllWindows()
```
